```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

We set the seed to 42 so that the data is reproducible and the graphs do not change with each execution we perform.

```python
np.random.seed(42)
```

Load the dataset and we will examine it.

```python
dataset = load_iris()
iris = pd.DataFrame(dataset.data, columns=dataset.feature_names)

iris['target'] = dataset.target
iris['species'] = iris['target'].map(
    dict(enumerate(dataset.target_names))
)
```

```python
iris.head()

```

```python
iris.describe()
```

# 1. Basic graphs

Create a bar chart and a histogram

```python
fig, ax = plt.subplots(figsize=(10,6))

mean = iris.groupby('species')['sepal length (cm)'].mean()

ax.bar(mean.index, mean.values)
plt.ylabel("Sepal length cm")
```

## **Conclusion**

As we can see in the graph, the **virginica** species has a greater sepal length than the setosa, although the versicolor also has a distance but is more reduced.

```python
fig, ax = plt.subplots(1,2, figsize=(16,8))

dist1 = iris['sepal length (cm)']
dist2 = iris['petal length (cm)']
nBins = 8

ax[0].hist(dist1, bins=nBins)
ax[0].set_title('sepal length (cm)')

ax[1].hist(dist2, bins=nBins)
ax[1].set_title('petal length (cm)')
```

## **Conclusion**

1. In the graph on the left, which represents the length of the sepal, we can see that the data is concentrated around the centre, showing a relatively uniform distribution, although there are some samples that extend towards lower values.

2. In contrast, in the graph on the right, corresponding to petal length, the distribution is less uniform: most of the data is concentrated towards lower values, indicating shorter lengths, while a few samples extend towards higher values.
In general, this reflects that petal length has greater asymmetry compared to sepal length.

# 2. Create a pie chart

Represent the species in a pie chart

```python
counts = iris['species'].value_counts()

fig, ax = plt.subplots(figsize=(10,5))

ax.pie(counts, labels=counts.index, autopct='%1.1f%%')
```

## **Conclusion**

The pie chart shows that the classes are perfectly balanced, with each species —Setosa, Versicolor, and Virginica— representing exactly 33.33% of the dataset.
This indicates that the dataset contains the same number of samples for each species, which is important for analyses and modeling, as it avoids class imbalance issues.
