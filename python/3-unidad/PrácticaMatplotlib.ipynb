```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generar datos simulados para 10 alumnos y 3 exámenes
np.random.seed(42) # para reproducibilidad
alumnos = [f"Alumno {i+1}" for i in range(10)]
examenes = [f"Examen {i+1}" for i in range(3)]
notas = np.random.randint(0, 100, size=(10, 3)) # Notas entre 50 y 99


# Crear el DataFrame
df_notas = pd.DataFrame(notas, index=alumnos, columns=examenes)


# Mostrar el DataFrame
print("DataFrame con las notas de los alumnos:")
print(df_notas)
```

# Ejercicio 1

Muestra una gráfica de barras y un histograma

```python
df_notas["Examen 1"].plot.bar()
```

```python
df_notas["Examen 1"].plot.hist(bins=5)
```

# Ejercicio 2

Muestra dos gráficas simultaneamente en la misma figura

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

df_notas['Examen 1'].plot(kind='bar', ax=axes[0], title='Gráfica bar')
df_notas['Examen 1'].plot(kind='hist', ax=axes[1], title='Gráfico hist')

plt.tight_layout()
```

# Ejercicio 3

En una misma figura, muestra en la gráfica de la izquierda la nota media por alumno ya la derecha una gráfica pastel donde se exprese el número de suspensos y aprobados

```python
from ast import Index
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

df_notas["media"] = df_notas.mean(axis = 1)

suspensos = df_notas[df_notas['media'] < 50.00]["media"]
df_notas["aprobado"] =  df_notas['media'] >= 50.00

labels = ["suspenso", "aprobados"]

axes[0].bar(df_notas.index ,df_notas["media"])
axes[0].set_title("Media de alumnos")
axes[0].tick_params(axis="x", rotation=45)

axes[1] = df_notas["aprobado"].value_counts().plot.pie()
axes[1].set_title("Media de aprobados")

```

# Ejercicio 3

Calcula la nota media de cada alumno y la nota media global del grupo, representa todas estas gráficas en una misma figura.

```python
df_notas["media"] = df_notas[['Examen 1', 'Examen 2', 'Examen 3']].mean(axis=1)

df_notas["global"] = df_notas["media"].mean()

df_notas["global"]

ax = df_notas['media'].plot(kind='bar', color='skyblue', label='media por alumno', figsize=(10, 6))
df_notas['global'].plot(kind='line', color='red', marker='o', label='Media global', ax=ax)

plt.title('Media individuales vs Globales')
plt.ylabel('Notas')
plt.xlabel('Alumnos')
plt.legend()
plt.xticks(rotation=45, ha='right')
```

# Ejercicio 4

Representar de forma apilada, las notas de los tres exámenes, solo para los alumnos cuya nota este suspensa

```python
alumnos_suspensos = df_notas[df_notas["aprobado"] == False]
alumnos_suspensos = alumnos_suspensos.drop(columns=["media","global"])
alumnos_suspensos.plot(kind="bar", stacked=True)
```
